# Modéliser le réel

- qu'est ce qu'est le réel ?

- qu'est ce qu'une mesure ?

- qu'est ce qu'un modèle ?

- qu'est ce qu'une erreur ?

## Le réel

**Exemple**

On considère la terre et le soleil

La fonction cible : ce qu'on cherche à estimer :

\begin{align}
h-r(t) = h_{real}(t) = f
\end{align}

```python
import numpy as np

import matplotlib.pyplot as plt

def h_real_function(X):

    return 60.0 + 20.0 * np.cos(2.5 * X * np.pi)

# Le temps "théorique" idéal régulier

t_real = np.linspace(0,1, 201)

# La hauteur réelle du soleil par rapport à l'horizon

h_real = h_real_function(t_real)

# on va afficher la fonction réelle

plt.figure()

plt.plot(t_real, h_real, label="Fonction cible")

plt.xlabel("Temps")

plt.ylabel("Hauteur dans le ciel en degrées")

plt.legend()

plt.show() 
```

## La mesure

On observe la course du soleil au dessus de l'horizon.

On effectue N mesures à plusieurs instants de la journée.

\begin{align}
T(t_0 \dots t_N)
\newline
h_m(T) = h_{measure}(T)
\end{align}

On a donc les données :

\begin{align}
T,h_m(T)
\end{align}

```python
N = 300

PRECISION_MEASURE = 5.0

# On a N mesures à des instants t aléatoires

t_measure = np.sort(np.random.rand(N))

# On a ensuite nos N mesure de hauteurs dans le ciel 

h_measure = h_real_function(t_measure) + np.random.randn(N) * PRECISION_MEASURE

# On affiche les données

plt.scatter(t_measure, h_measure, label="Mesure")

plt.legend()

plt.xlabel("Temps")

plt.ylabel("Hauteur dans le ciel")

plt.show()
```

Cette mesure on peut l'écrire sous la forme : 

\begin{align}
h_m(t)
\end{align}

## Le modèle

On cherche un modèle de la hauteur du soleil dans le ciel

\begin{align}
h_e(t) = h_{estimate}(t)
\end{align}

Pour trouver le bon modèle on doit savoir estier l'erreur entre le modèle et les donner pour pouvoir minimiser cette erreur.

\begin{align}
Err(t) = E \left[ (h_m(t) - h_e(t) )^2 \right]
\end{align}

On peut montrer que cette erreur se décompose sous la forme suivante : 

\begin{align}
Err(t) = \sigma_{\epsilon}^2(t) + V \left[ h_e(t) \right] + Bias^2(t)
\end{align}

### Régression linéaire

On choisit de modéliser les donneés par un droite:

\begin{align}
h_e(t) = a \times t + b
\end{align}

```python
from sklearn.linear_model import LinearRegression

linear_regression = LinearRegression()
linear_regression.fit(t_measure[:, np.newaxis], h_measure)
```

```python
h_e_linear = linear_regression.predict(t_real[:, np.newaxis])

plt.scatter(t_measure, h_measure, label="Mesure")
plt.plot(t_real, h_e_linear, label="Modèle linéaire", color="red")

plt.xlabel("temps")
plt.ylabel("hauteur")
plt.legend()
```

### Polynôme

Pour rappel:

\begin{align}
y = a0
\end{align}

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

d_degree = 4

#on prepare le "solver" de regression polynomiale
polynomial_features = PolynomialFeatures(degree=d_degree, include_bias=True)
polynomial_regression = Pipeline(
    [
        ("poly", polynomial_features),
        ("regression", LinearRegression())
    ]
)

# on l'entraine sur les mesures
polynomial_regression.fit(t_measure[:, np.newaxis], h_measure)

#on va afficher le resultat: notre model polynomial
h_e_poly_2 = polynomial_regression.predict(t_real[:, np.newaxis])

plt.scatter(t_measure, h_measure, label="Mesure")
plt.plot(t_real, h_e_linear, label="Modèle linéaire", color="red")

plt.plot(t_real, h_e_poly_2, label="Modèle polynomial d="+str(d_degree), color="blue")

plt.xlabel("temps")
plt.ylabel("hauteur")
plt.legend()
plt.show()
```

### TP

- faire varier le degré du polynome dans une boucle
- Afficher tout cela proprement
- Calculer le score de cross validation pour les différents degrés et l'afficher dans un graphique


```python
from sklearn.model_selection import cross_val_score
from pprint import pprint

i = 1

while i <= 20:

    d_degree = i

    #on prepare le "solver" de regression polynomiale
    polynomial_features = PolynomialFeatures(degree=d_degree, include_bias=True)
    polynomial_regression = Pipeline(
        [
            ("poly", polynomial_features),
            ("regression", LinearRegression())
        ]
    )

    # on l'entraine sur les mesures
    polynomial_regression.fit(t_measure[:, np.newaxis], h_measure)

    #on va afficher le resultat: notre model polynomial
    h_e_poly_2 = polynomial_regression.predict(t_real[:, np.newaxis])
    
    # calcule du score de cross validation TODO
    pprint(cross_val_score(polynomial_regression, t_measure[:, np.newaxis], h_measure[:, np.newaxis], cv=10))
    
    plt.scatter(t_measure, h_measure, label="Mesure")
    plt.plot(t_real, h_e_linear, label="Modèle linéaire", color="red")

    plt.plot(t_real, h_e_poly_2, label="Modèle polynomial d="+str(d_degree), color="blue")

    plt.xlabel("temps")
    plt.ylabel("hauteur")
    plt.ylim(20, 100)
    plt.legend()
    plt.show()
    i += 1
```

**Mean squared error score**

\begin{align}
MSE(y, g) = \frac{1}{N} \sum_{i=0}^N (y_i - g_i)^2
\end{align}

**Explained variance**

\begin{align}
EV(y, g) = 1 - \frac{Var[y - g]}{var[y]}
\end{align}

```python
i = 1

score_means = list()
score_stds = list()

degrees = (4, 5, 6, 7, 8, 9)

for degree in degrees:

    d_degree = degree

    #on prepare le "solver" de regression polynomiale
    polynomial_features = PolynomialFeatures(degree=d_degree, include_bias=True)
    polynomial_regression = Pipeline(
        [
            ("poly", polynomial_features),
            ("regression", LinearRegression())
        ]
    )

    # on l'entraine sur les mesures
    polynomial_regression.fit(t_measure[:, np.newaxis], h_measure)

    #on va afficher le resultat: notre model polynomial
    h_e_poly_2 = polynomial_regression.predict(t_real[:, np.newaxis])
    
    # calcule du score de cross validation TODO
    cross_val = cross_val_score(
        polynomial_regression, 
        t_measure[:, np.newaxis], 
        h_measure[:, np.newaxis], 
        scoring="neg_mean_squared_error",
        cv=10
    )
    score_means.append(-cross_val.mean())
    score_stds.append(cross_val.std())
    pprint(cross_val.std())
    score_mean = cross_val.mean()
    pprint("degree "+str(d_degree)+" : "+str(-score_mean))
    
plt.errorbar(degrees, score_means, np.array(score_stds))

plt.xlabel('degree')
plt.ylabel('score rate')
#plt.ylim(0, 350)

plt.show()
```

```python
N_TEST = 40
best_degree = 6
polynomial_features = PolynomialFeatures(degree=best_degree, include_bias=True)
best_estimator = Pipeline(
    [
        ("poly", polynomial_features),
        ("regression", LinearRegression())
    ]
)
best_estimator.fit(t_measure[:, np.newaxis], h_measure)

t_test = np.random.uniform(1.0, 2.0, size=N_TEST)

h_test = h_real_function(t_test) + np.random.randn(N_TEST) * PRECISION_MEASURE
h_best_on_test = best_estimator.predict(t_test[:, np.newaxis])
print("r2 score")
print(best_estimator.score(t_test[:, np.newaxis], h_test))

ax = plt.figure().gca()
plt.scatter(t_measure, h_measure, label="Learning sample")
plt.scatter(t_test, h_test, label="new test data")
plt.scatter(t_test, h_best_on_test, label="best poly on test data")
ax.set_yscale("log")
plt.ylabel("hauteur: y")
plt.xlabel("Temps: x")
plt.legend()
plt.show()
```
