# Classifier Naive Bayes

Technique de classification basée sur le théorème de Bayes.
On a des mesures X et des classes C. Chaque mesure appartien à une classe.

Le thèoreme de Bayes nous dit:
\begin{align}
P(C|X) = \frac{P(X|C) \times P(C)}{P(X)}
\end{align}

avec 

- $P(C|X)$ la probabilité de la classe C sachant X (probabilité postérieure)
- $P(C)$ la probabilité a priori de la classe C
- $P(X)$ la probabilité a priori de la classe X
- $P(X|C)$ la probabilité de la classe X sachant C

Le classifier NB fait l'hypothèse de l'indépendance des mesures.
Une bannane est jaune, allongée et sucré: NB va 

Ca qui est probablement faux en général d'où le non "Naive"

Les avantages de cette Hypothèse:
- rapide
- simple à comprendre
- marche bien en haute dimension

Les inconvénients
- on peut surment faire mieux
- marche mal avec peu de dimension

\begin{align}
P(X) = P(x_1) P(x_2) \dots P(x_n)
\end{align}

et donc

\begin{align}
P(C|X) = \frac{P(x_1|C) P(x_2|C) \dots P(x_n|C) P(C}{P(X)}
\end{align}

## Exemple

**les mesures**

|taille|gout|couleur|classe|
|-|-|-|-|
|long|sucré|jaune|banane|
|rond|acide|autre|orange|
|long|sucré|autre|banane|
|long|acide|autre|autre|

**les agrégats**

|classe|long|rond|sucré|acide|jaune|autre|
|-|-|-|-|-|-|-|
|banane|500|0|-|-|-|-|

**Les probabilités**

Probabilités des classes : $P(C)$
\begin{align}
P(Banane) = \frac{500}{1000} = 0.5
\newline
\dots
\end{align}

Probabilités des cxaractéristiques : $P(X)$
\begin{align}
P(Long) = \frac{500}{1000} = 0.5
\newline
\dots
\end{align}

**Un Observation**
On observe un fruit rond, sucré et jaune. **Quelle est sa classe?**
\begin{align}
P(Banane|X) = \frac{P(Rond|Banane) \times P(Sucré|Banane) \times P(jaune|Banane)}{P(rond) \times P(sucré) \times P(jaune)}
\newline
P(Orange|X) = \frac{P(Rond|Orange) \times P(Sucré|Orange) \times P(jaune|Orange)}{P(rond) \times P(sucré) \times P(jaune)}
\end{align}
